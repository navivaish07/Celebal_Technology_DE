```python
price_columns = ["initial_price", "final_price"]
for col in price_columns:
        data[col] = (
        data[col]
        .astype(str)
        .str.replace(r'[^0-9.]', '', regex=True)
    )
        data[col] = pd.to_numeric(
        data[col],
        errors='coerce'
    )
```

```python
data[["initial_price", "final_price"]].head()
```

```python
data["discount"]= data["discount"].fillna(data["discount"].mean())
```

```python
text_cols = [
    "seller_name",
    "seller_information",
    "what_customers_said",
    "videos",
    "variations"
]

for col in text_cols:
    data[col] = data[col].fillna("Not Available")
```

```python
print("Before:", data.shape)
data.drop_duplicates(inplace=True)
print("After:", data.shape)
```

```python
data["price_difference"] = (
    data["initial_price"] -
    data["final_price"]
)
```

```python
data[["initial_price",
     "final_price",
     "price_difference"]].sample(20)
```

```python
data["popularity_score"] = ( data["rating"] * data["ratings_count"])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data["final_price"],bins=20)
plt.title("Distribution of Final Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data["rating"],bins=10)
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="rating",y="final_price",data=data)
plt.title("Rating vs Final Price")
plt.show()
```

```python
avg_price = (
    data.groupby("category")
      ["final_price"]
      .mean()
      .sort_values(ascending=False)
      .tail(10))
plt.figure(figsize=(12,6))
avg_price.plot(kind="bar")
plt.title("Average Price by Category")
plt.show()
```

```python
category_counts = (
    data["category"]
    .value_counts()
    .head(10)
)
print(category_counts)
```

```python
plt.figure(figsize=(12,6))
category_counts.plot(kind="bar")
plt.title("Top 10 Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```
